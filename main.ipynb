```python
# Get the urls for one single paper
# of one single institution

# TODO: the authors above also include collaborators from other unis

def get_list_urls_paper(paper):
    list_urls_authors_one_paper = []
    inner_paper = paper.get('metadata').get('authors')
    for i in range(len(inner_paper)):
        url = inner_paper[i].get('record')
        if url is not None:
            url = url.get('$ref')
            list_urls_authors_one_paper.append(url)
    return list_urls_authors_one_paper

get_list_urls_paper(papers_of_one_institution[9])
```

```python
# Get list of lists for all the papers
# of one institution 
def get_list_urls_institution(papers_of_one_institution):
    papers_of_one_institution = papers_of_one_institution
    list_urls_institution = []
    for i in range(len(papers_of_one_institution)):
        list_urls = get_list_urls_paper(papers_of_one_institution[i])
        list_urls_institution.append(list_urls)
    return list_urls_institution

authors_ilorin = get_list_urls_institution(papers_of_one_institution)
authors_ilorin
```

```python
# Flatten and remove duplicates while preserving order
flat_authors_ilorin = []
for paper in authors_ilorin:
    for url in paper:
        if url not in flat_authors_ilorin:
            flat_authors_ilorin.append(url)

flat_authors_ilorin
```

```python
flat_authors_ilorin_clean = [url.replace('/api', '') for url in flat_authors_ilorin]
flat_authors_ilorin_clean
```
